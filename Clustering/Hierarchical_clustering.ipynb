# Hierarchical Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.loc[:, ['Annual Income (k$)', 'Spending Score (1-100)']]
```

## Using the dendrogram to find the optimal number of clusters

```
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Eucludian Distances')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
hc.fit(X)
```

## Visualising the clusters

```
X = np.array(X)
plt.scatter(X[hc.labels_ == 0, 0], X[hc.labels_ == 0, 1], s=20, c='red', label='Cluster 1')
plt.scatter(X[hc.labels_ == 1, 0], X[hc.labels_ == 1, 1], s=20, c='blue', label='Cluster 2')
plt.scatter(X[hc.labels_  == 2, 0], X[hc.labels_ == 2, 1], s=20, c='green', label='Cluster 3')
plt.scatter(X[hc.labels_  == 3, 0], X[hc.labels_ == 3, 1], s=20, c='yellow', label='Cluster 4')
plt.scatter(X[hc.labels_  == 4, 0], X[hc.labels_ == 4, 1], s=20, c='purple', label='Cluster 5')
plt.legend()

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

## Notes
The customers with high income (green and red) are the ones we actually want to target.
The yellow ones should be protected and sent less advertising for ethical reasons.

---

We also Notice that when only use 3 clusters, we get less accurate predictions, but the final decision is the same. While 5 is the optimal number of clusters, 3 is also not bad.
