Importing and Loading Spacy

```python
spacy.cli.download("en_core_web_sm")
import spacy

nlp = spacy.load("en_core_web_sm")
```

The text to be processed

```python
my_sentence = "Hello, I'm the french librarian gazing a the stars above the ringing"
```

Creating the Tokenization function

```python
def return_token_text(sentence):
    # Divide the sentence into tokens
    doc = nlp(sentence)
    # Return the text of the tokens
    return [token.text for token in doc]
```

```python
my_tok_text = return_token_text(my_sentence)
print("Tokenizing...")
print(my_tok_text)
```

Importing stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stopWords = set(stopwords.words('french'))
```

Ignoring the stopwords

```python
clean_words = []


def remove_stopwords(sentence):
    for token in return_token_text(sentence):
        if token not in stopWords:
            clean_words.append(token)
```

```python
remove_stopwords(my_sentence)
print("Removing stopwords...")
print(clean_words)
```

Importing SnowballStemmer

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='french')
```

Stemming

```python
def return_stem(sentence):
    doc = nlp(sentence)
    return [stemmer.stem(s.text) for s in doc]
```

```python
my_stem = return_stem(my_sentence)
print("Stemming")
print(my_stem)
```
