# Normal Distribution

### Why the check should we care about the normal distribution of our data?

**Assumptions of models**<br>
Some models such as linear regression assume that the data follows a normal distribution. If the data deviates significantly from a normal distribution, their validty may be compromised. 

**Normalization**<br>
If the data does not follow a normal distribution, it may be necessary to apply transformations or normalization techniques to make it conform to the assumptions of the models or to improve their performance.

**Model Performances**<br>
Models based on normality assumptions may perform better when the data follows a normal distribution. Deviations from normality can affect the model's accuracy, prediction, and generalization capabilities. 

**Outliers and influential observations Identification**<br>
Outliers are data points that significantly deviate from the majority of the data.
When the data follows a normal distribution, it is often easier to identify outliers and influential observations. Outliers can be identified based on their deviation from the mean or other measures of central tendency. In a normal distribution, extreme values are less likely to occur, making it easier to distinguish them from the majority of the data.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Loading the data

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = iris.data
sepal_length = data[:, 0]
sepal_width = data[:, 1]
petal_length = data[:, 2]
petal_width = data[:, 3]
```

## Plotting the distribution of the features

### Kernel Density Estimate (KDE) plot
The dsitribution of the data is not a bell-shaped curve, so the data is not normally distributed.  

```python
def plot_distribution(data, feature):
    """ Create a histogram with ernel density estimate (KDE) plot"""
    sns.histplot(sepal_length, kde=True)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Distribution {feature}')
```

```python
plt.figure(figsize=(9,11))

plt.subplot(221)
plot_distribution(sepal_length, 'Sepal Length')

plt.subplot(222)
plot_distribution(sepal_width, 'Sepal Width')

plt.subplot(223)
plot_distribution(petal_length, 'Petal Length')

plt.subplot(224)
plot_distribution(petal_width, 'Petal Width')

plt.show()
```

### Q-Q Plot
A Q-Q plot (quantile-quantile plot) is a graphical tool used to assess if a given dataset follows a particular theoretical distribution, such as the normal distribution. 
The resulting plot will show the data points plotted against the expected quantiles of the normal distribution. If the data points closely follow the diagonal line, it suggests that the data follows the normal distribution. Deviations from the line indicate departures from normality as in the following plot.

```python
stats.probplot(sepal_length, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

## Checking if the data is normally distributed

### Features dsitribution - Shapiro-Wilk test

```python
from sklearn.datasets import load_iris
from scipy import stats

iris = load_iris()
alpha = 0.05

for feature_name in iris.feature_names:
    feature_data = iris.data[:, iris.feature_names.index(feature_name)]
    stat, p = stats.shapiro(feature_data)

    if p > alpha:
        print(f"The {feature_name} feature follows a normal distribution.")
    else:
        print(f"The {feature_name} feature does not follow a normal distribution.")
```
